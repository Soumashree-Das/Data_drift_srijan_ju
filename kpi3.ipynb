```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_data.csv")
```

```python
df.head()
```

```python
df['Time of Day'].unique()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Day of Week'] = df['Date'].dt.day_name()

# Optional: order Time of Day and Day of Week
time_order = ['Morning', 'Afternoon', 'Evening', 'Night']
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Group by Time of Day and Day of Week to count accidents
grouped = df.groupby(['Time of Day', 'Day of Week']).size().reset_index(name='Accidents')

# Pivot for heatmap
pivot_df = grouped.pivot(index='Time of Day', columns='Day of Week', values='Accidents')
pivot_df = pivot_df.reindex(index=time_order, columns=day_order)

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_df, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Accident Patterns by Time of Day and Day of Week', fontsize=14)
plt.xlabel('Day of Week')
plt.ylabel('Time of Day')
plt.show()

# Weekday vs Weekend classification
df['Weekend'] = df['Day of Week'].isin(['Saturday', 'Sunday'])

# Group and count for barplot
weekend_grouped = df.groupby(['Weekend', 'Time of Day']).size().reset_index(name='Accidents')

# Plot bar chart
plt.figure(figsize=(10, 5))
sns.barplot(data=weekend_grouped, x='Time of Day', y='Accidents', hue='Weekend', order=time_order)
plt.title('Weekday vs Weekend Accidents by Time of Day', fontsize=14)
plt.xlabel('Time of Day')
plt.ylabel('Number of Accidents')
plt.legend(title='Weekend', labels=['Weekday', 'Weekend'])
plt.show()
```
